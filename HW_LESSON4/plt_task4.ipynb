### Тема “Визуализация данных в Matplotlib”

```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import pandas as pd 
creditcard_df = pd.read_csv('creditcard.csv', delimiter=',') # открыть датасет из корневой папки Jupiter
creditcard_df.iloc[:10] # вывод десяти первых строк датасета
```

```python
count_class=creditcard_df['Class'].value_counts() # количество наблюдений для каждого значения целевой переменной Class
print(count_class)
```

```python
count_class.plot(kind="bar")
plt.style.use('fivethirtyeight')
plt.show()
```

```python
count_class.plot(kind="bar", logy=True)
plt.style.use('fivethirtyeight')
plt.show()
```

```python
plt.hist(creditcard_df.loc[creditcard_df['Class']==0,'V1'], color='grey', bins=20,alpha=0.5, density=True)
plt.hist(creditcard_df.loc[creditcard_df['Class']==1,'V1'], color='red', alpha=0.5, bins=20, density=True)

title_font = {
"fontsize": 16,
"fontweight": "bold",
"color": "#000080",
"family": "serif",
}
plt.title("Гистограмма", fontdict=title_font)
plt.legend(labels=["Class 0", "Class 1"])

label_font = {
"fontsize": 12,
"family": "monospace",
"color": "#000089",
}
plt.xlabel("Class", fontdict=label_font)
plt.axis([-35, 5, 0, 0.25])
plt.show()
```
