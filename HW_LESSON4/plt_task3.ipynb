```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from pylab import rcParams
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

# Тема “Визуализация данных в Matplotlib”

```python
x = np.linspace(-3, 3, 51)
print(x)
len(x)
```

```python
y1 = x ** 2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
y1, y2, y3, y4
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
# ax1, ax2, ax3, ax4 = ax.flatten()

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.5)

ax[0, 0].plot(x, y1, color="#00FF00")
ax[0, 0].set_title("График у1")
ax[0, 0].grid(axis = 'x',
        color = 'blue',    #  цвет линий сетки
        linewidth = 0.5,    #  толщина линии сетки
        linestyle = '--')    #  начертание линии сетки
ax[0, 0].set_xlim([-5, 5])

ax[0, 1].plot(x, y2, color="#4B0082")
ax[0, 1].set_title("График у2")
ax[0, 1].grid(axis = 'y',
        color = 'green',    #  цвет линий сетки
        linewidth = 0.3,    #  толщина линии сетки
        linestyle = '--')    #  начертание линии сетки

ax[1, 0].plot(x, y3, color="#FF0000")
ax[1, 0].set_title("График у3")
ax[1, 0].grid(axis = 'both',
        color = 'grey',    #  цвет линий сетки
        linewidth = 0.8,    #  толщина линии сетки
        linestyle = '--')    #  начертание линии сетки

ax[1, 1].plot(x, y4, color="#000000")
ax[1, 1].set_title("График у4")
ax[1, 1].grid(axis = 'both',
        color = 'OrangeRed',    #  цвет линий сетки
        linewidth = 1,    #  толщина линии сетки
        linestyle = '--')    #  начертание линии сетки
ax[1, 1].set_ylim([-2, 2])

```
